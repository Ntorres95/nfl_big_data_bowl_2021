```python
import pandas as pd
```

```python
df = pd.read_csv("Resources/plays.csv")
df.head()
```

```python
df = df.drop(columns=['playDescription','possessionTeam', 'playType', 
                        'yardlineSide', 'gameClock', 'penaltyCodes', 
                        'penaltyJerseyNumbers',
                     'personnelO', 'personnelD', 'typeDropback', 'offenseFormation'])
```

```python
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')

# Drop the null rows
df = df.dropna()
df.head()
```

```python
# Set features. This will also be used as your x values.
selected_features = df[['down', 'yardsToGo', 'isDefensivePI']]
```

```python
from sklearn.model_selection import train_test_split
y = df["passResult"]
X = df.drop(columns=["passResult"])

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(X_train_scaled, y_train)
```

```python
print(f"Training Data Score: {model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model.score(X_test_scaled, y_test)}")
```
